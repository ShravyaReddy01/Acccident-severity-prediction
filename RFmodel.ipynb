Import Data

```python
import pandas as pd
df=pd.read_csv("Data.csv", encoding="ISO-8859-1")
df.head(10)

```

Data Preprocessing


```python
#checking for null values
print(df.isnull().sum())

#check for duplicate columns
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
print(df.columns)
print(df.shape)
```

```python
#converting 'time' column to datetime
df['Time']=pd.to_datetime(df['Time'])
df['Time'].dt.hour

#extracting hour from time column
df['hour']=df['Time'].dt.hour
df.drop('Time', axis=1, inplace=True)
df.head()
```

```python
#Dropping empty columns
df.drop('Unnamed: 32',axis=1)
```

```python
#Getting the null value columns
df.isna().sum()[df.isna().sum()!=0]
```

```python
#dropping columns that can cause imbalance while imputation and irrelevant factors
df=df.drop(columns=['Educational_level','Sex_of_driver','Pedestrian_movement','Cause_of_accident','Casualty_class','Vehicle_driver_relation','Owner_of_vehicle','Service_year_of_vehicle','Defect_of_vehicle','Area_accident_occured','Types_of_Junction','Vehicle_movement','Work_of_casuality','Fitness_of_casuality','Unnamed: 32','Sex_of_casualty','Number_of_casualties','Number_of_vehicles_involved','Type_of_collision','Light_conditions','Road_surface_conditions','Road_surface_type','Lanes_or_Medians','Road_allignment','Age_band_of_casualty','Casualty_severity'])
```

```python
df.head()
```

```python
#imputing columns with null values
impute_cols = [x for x in df.isna().sum()[df.isna().sum() != 0].index.tolist()]

for feature in impute_cols:
    mode_series = df[feature].mode()
    
    # Check if mode_series is empty (no mode found)
    if not mode_series.empty:
        mode_value = mode_series[0]
        df[feature].fillna(mode_value, inplace=True)
```

```python
df.isna().sum()
```

Encoding

```python
import numpy as np
def ordinal_encoder(df,feats):
    for feat in feats:
        feat_val=list(np.arange(df[feat].nunique()))
        feat_key=list(df[feat].sort_values().unique())
        feat_dict=dict(zip(feat_key, feat_val))
        df[feat]=df[feat].map(feat_dict)
    return df
df=ordinal_encoder(df,df.drop(['Accident_severity'],axis=1).columns)
df.shape
```

```python
df.head()
```

Machine learning


```python
import optuna
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
x=df.drop('Accident_severity',axis=1)
y=df['Accident_severity']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Assuming you have multiclass classification
Accuracy = accuracy_score(y_test, y_pred)
Precision = precision_score(y_test, y_pred, average='weighted')
Recall = recall_score(y_test, y_pred, average='weighted')
F1_score = f1_score(y_test, y_pred, average='weighted')

# Print the results
print("Accuracy:", Accuracy)
print("Precision:", Precision)
print("Recall:", Recall)
print("F1 Score:", F1_score)
```
